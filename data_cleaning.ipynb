```python
import pandas as pd

# 读取 CSV 数据
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')

# 预览前几行
print(df.head())

# 查看字段信息
print(df.info())
```

```python
# 将订单日期转换为 datetime 格式
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# 创建订单月份字段（格式：YYYY-MM）
df['Month'] = df['ORDERDATE'].dt.to_period('M').astype(str)

# 可选：添加年份列
df['Year'] = df['ORDERDATE'].dt.year

# 计算每一行订单的总销售额
df['TOTAL_SALES'] = df['QUANTITYORDERED'] * df['PRICEEACH']

# 预览新增的字段
df[['ORDERNUMBER', 'ORDERDATE', 'Month', 'TOTAL_SALES']].head()
```

```python
monthly_sales = df.groupby('Month')['TOTAL_SALES'].sum().reset_index()
monthly_sales['TOTAL_SALES'] = monthly_sales['TOTAL_SALES'].astype(float)
monthly_sales.to_csv('monthly_sales.csv', index=False, float_format="%.2f", encoding='utf-8-sig')

country_sales = df.groupby('COUNTRY')['TOTAL_SALES'].sum().reset_index()
country_sales['TOTAL_SALES'] = country_sales['TOTAL_SALES'].astype(float)
country_sales.to_csv('country_sales.csv', index=False, float_format="%.2f", encoding='utf-8-sig')

top_products = df.groupby('PRODUCTLINE')['TOTAL_SALES'].sum().reset_index()
top_products['TOTAL_SALES'] = top_products['TOTAL_SALES'].astype(float)
top_products.to_csv('top_products.csv', index=False, float_format="%.2f", encoding='utf-8-sig')
```
